```python
import pandas as pd
import re
```

```python
df = pd.read_csv("combined_dpr.csv")
print(df.shape)
df.head(5)
```

```python
# drop duplicates & missing values
df = df.drop_duplicates(subset=["Comment"])
df = df.dropna(subset=["Comment"])
```

```python
# remove non-textual noise (URLs, mentions, hashtags)
def clean_comment(text):
    text = str(text)
    text = re.sub(r"http\S+|www\S+|https\S+", "", text) # urls
    text = re.sub(r"@\w+", "", text)  # mentions
    text = re.sub(r"#\w+", "", text)  # hashtags
    text = re.sub(r"\s+", " ", text).strip()  # normalize whitespace
    return text

df["clean_comment"] = df["Comment"].apply(clean_comment)

# lowercase for consistency
df["clean_comment"] = df["clean_comment"].str.lower()

# 4. remove very short or empty comments
df = df[df["clean_comment"].str.len() > 2]

# save cleaned dataset
df.to_csv("cleaned_dpr.csv", index=False)

print("Before cleaning:", len(pd.read_csv("combined_dpr.csv")))
print("After cleaning:", len(df))
```
